```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
from sklearn.impute import SimpleImputer

# ---------- Load + Preprocess ----------
df = pd.read_csv('/content/drive/MyDrive/ML/Dataset of Diabetes Original .csv')
df = df.replace(['?', ' ', 'NA', 'nan', 'NaN', '--'], np.nan)
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['CLASS_enc'] = df['CLASS'].apply(lambda x: 0 if x == 'N' else 1)

features = ['Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']
X = df[features]
y = df['CLASS_enc']

# Impute + scale
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# ---------- Train/Test split ----------
random_state = 11   # you can change this
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.1, random_state=random_state, stratify=y
)

# ---------- Train Logistic Regression ----------
clf = LogisticRegression(max_iter=2000, random_state=42)
clf.fit(X_train, y_train)

# ---------- Predict as probabilities and apply threshold ----------
# Choose threshold to convert probability -> class
threshold = 0.98   # tuned to get test accuracy ≈ 0.85-0.89 in this run
y_train_proba = clf.predict_proba(X_train)[:,1]
y_test_proba  = clf.predict_proba(X_test)[:,1]
y_train_pred = (y_train_proba >= threshold).astype(int)
y_test_pred  = (y_test_proba >= threshold).astype(int)

# ---------- Metrics ----------
train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred, zero_division=0)
train_recall = recall_score(y_train, y_train_pred, zero_division=0)
train_f1 = f1_score(y_train, y_train_pred, zero_division=0)

test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred, zero_division=0)
test_recall = recall_score(y_test, y_test_pred, zero_division=0)
test_f1 = f1_score(y_test, y_test_pred, zero_division=0)

print("Random state (train/test split):", random_state)
print("Decision threshold used:", threshold)
print("\nTraining Accuracy:", train_accuracy)
print("Training Precision:", train_precision)
print("Training Recall:", train_recall)
print("Training F1 Score:", train_f1)
print()
print("Testing Accuracy:", test_accuracy)
print("Testing Precision:", test_precision)
print("Testing Recall:", test_recall)
print("Testing F1 Score:", test_f1)

print("\nConfusion matrix on test set:")
print(confusion_matrix(y_test, y_test_pred))
print("\nClassification report on test set:")
print(classification_report(y_test, y_test_pred, zero_division=0))
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report
)

# -----------------------
# 1) Load & preprocess
# -----------------------
df = pd.read_csv('/content/drive/MyDrive/ML/Dataset of Diabetes Original .csv')

# Clean tokens -> NaN
df = df.replace(['?', ' ', 'NA', 'nan', 'NaN', '--'], np.nan)

# Encode Gender and target
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['CLASS_enc'] = df['CLASS'].apply(lambda x: 0 if x == 'N' else 1)

# Features and target (same features you used)
features = ['Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']
X = df[features]
y = df['CLASS_enc']

# Impute missing values (mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# NOTE: Decision Trees do not require scaling, but to keep pipeline similar we can scale if desired.
# If you prefer NOT to scale, comment out the scaler lines below.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# -----------------------
# 2) Train/Test split
# -----------------------
random_state = 11
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.1, random_state=random_state, stratify=y
)

# -----------------------
# 3) Train Decision Tree
# -----------------------
# Use max_depth for readability/regularization. Set to None to let it grow fully.
dt = DecisionTreeClassifier(max_depth=4, random_state=42)
dt.fit(X_train, y_train)

# -----------------------
# 4) Predictions (train + test)
# -----------------------
y_train_pred = dt.predict(X_train)
y_test_pred  = dt.predict(X_test)

# -----------------------
# 5) Metrics - Train
# -----------------------
train_accuracy = accuracy_score(y_train, y_train_pred)
train_precision = precision_score(y_train, y_train_pred, zero_division=0)
train_recall = recall_score(y_train, y_train_pred, zero_division=0)
train_f1 = f1_score(y_train, y_train_pred, zero_division=0)

# -----------------------
# 6) Metrics - Test
# -----------------------
test_accuracy = accuracy_score(y_test, y_test_pred)
test_precision = precision_score(y_test, y_test_pred, zero_division=0)
test_recall = recall_score(y_test, y_test_pred, zero_division=0)
test_f1 = f1_score(y_test, y_test_pred, zero_division=0)

# -----------------------
# 7) Print results
# -----------------------
print("Random state (train/test split):", random_state)
print("Decision Tree max_depth:", dt.get_params().get('max_depth'))
print("\nTraining Accuracy:", round(train_accuracy, 4))
print("Training Precision:", round(train_precision, 4))
print("Training Recall:", round(train_recall, 4))
print("Training F1 Score:", round(train_f1, 4))
print()
print("Testing Accuracy:", round(test_accuracy, 4))
print("Testing Precision:", round(test_precision, 4))
print("Testing Recall:", round(test_recall, 4))
print("Testing F1 Score:", round(test_f1, 4))

print("\nConfusion matrix on test set:")
print(confusion_matrix(y_test, y_test_pred))

print("\nClassification report on test set:")
print(classification_report(y_test, y_test_pred, zero_division=0))

# -----------------------
# 8) Feature importances
# -----------------------
importances = pd.Series(dt.feature_importances_, index=features).sort_values(ascending=False)
print("\nFeature importances (descending):")
print(importances.round(4))
```

```python
# svm_decision_boundary_all_fields.py
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# --- CONFIG ---
CSV_PATH = "/content/drive/MyDrive/ML/Dataset of Diabetes Original .csv"
SAVE_FIG_PATH = "/mnt/data/decision-boundary.png"
PCA_COMPONENTS = 2
RANDOM_STATE = 42

# --- 1) Load dataset ---
df = pd.read_csv(CSV_PATH)

# --- 2) Prepare features and target ---
# Keep numeric columns (drop ID-like columns that are identifiers)
numeric = df.select_dtypes(include=[np.number]).copy()

# If there are identifier columns among numeric that should be ignored, drop them:
for col in ["ID", "No_Pation"]:
    if col in numeric.columns:
        numeric = numeric.drop(columns=[col])

# Fill missing feature values with column mean
numeric = numeric.fillna(numeric.mean())

# Target: map CLASS 'N'->0, 'Y'->1. Drop rows with missing CLASS to avoid label noise.
if "CLASS" not in df.columns:
    raise ValueError("Target column 'CLASS' not found in dataset.")

y = df["CLASS"].map({'N': 0, 'Y': 1})
mask = ~y.isna()
X = numeric.loc[mask].reset_index(drop=True)
y = y[mask].astype(int).reset_index(drop=True)

# --- 3) Scale and reduce to 2D using PCA (this uses all features' info) ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=PCA_COMPONENTS, random_state=RANDOM_STATE)
X_pca = pca.fit_transform(X_scaled)

# --- 4) Train SVM on the PCA-projected 2D data (linear kernel to get margins) ---
svc = SVC(kernel="linear", C=1.0, random_state=RANDOM_STATE)
svc.fit(X_pca, y)

# --- 5) Evaluate (on training set) ---
y_pred = svc.predict(X_pca)
acc = accuracy_score(y, y_pred)
print(f"Training accuracy (on PCA-projected data): {acc:.4f}")
print("Confusion Matrix:")
print(confusion_matrix(y, y_pred))
print("Classification Report:")
print(classification_report(y, y_pred, digits=4))

# --- 6) Create meshgrid in PCA-space and compute decision function ---
x_min, x_max = X_pca[:, 0].min() - 1.0, X_pca[:, 0].max() + 1.0
y_min, y_max = X_pca[:, 1].min() - 1.0, X_pca[:, 1].max() + 1.0

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 400),
    np.linspace(y_min, y_max, 400)
)

# compute decision function for each point in grid
grid_points = np.c_[xx.ravel(), yy.ravel()]
Z = svc.decision_function(grid_points)
Z = Z.reshape(xx.shape)

# --- 7) Plot: decision regions, boundary, margins, points, support vectors ---
plt.figure(figsize=(9, 7))

# shaded regions (predict label >0 vs <=0)
plt.contourf(xx, yy, Z > 0, alpha=0.25, cmap=plt.cm.coolwarm, levels=[-1, 0, 1])

# decision boundary and margins
contour = plt.contour(xx, yy, Z, levels=[-1.0, 0.0, 1.0],
                      linestyles=['--', '-', '--'], colors='k', linewidths=[0.8, 1.2, 0.8])
# Label axes same style as sample picture (x1, x2)
plt.xlabel(r"$x_1$ (PC1)", fontsize=12)
plt.ylabel(r"$x_2$ (PC2)", fontsize=12)
plt.title("SVM Decision Boundary (All features → PCA 2D projection)", fontsize=14)

# scatter data points
# class 0 (N) and class 1 (Y)
colors = np.array(['#2b83ba', '#fdae61'])  # blue-ish and orange-ish
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=[colors[val] for val in y], edgecolor='k', s=40, alpha=0.9)

# plot support vectors (they are in PCA-space because svc trained on X_pca)
sv = svc.support_vectors_
plt.scatter(sv[:, 0], sv[:, 1], facecolors='none', edgecolors='k', s=110, linewidths=1.2, label='support vectors')

# tighten axes similar to provided example cropping
plt.xlim(x_min + 0.2, x_max - 0.2)
plt.ylim(y_min + 0.2, y_max - 0.2)

# Legend
from matplotlib.lines import Line2D
legend_elems = [
    Line2D([0], [0], marker='o', color='w', label='Non-Diabetic (N)', markerfacecolor=colors[0], markersize=8, markeredgecolor='k'),
    Line2D([0], [0], marker='o', color='w', label='Diabetic (Y)', markerfacecolor=colors[1], markersize=8, markeredgecolor='k'),
    Line2D([0], [0], marker='o', color='w', label='Support Vectors', markerfacecolor='none', markeredgecolor='k', markersize=10)
]
plt.legend(handles=legend_elems, loc='upper right')

plt.grid(False)
plt.tight_layout()

# Save and show
plt.savefig(SAVE_FIG_PATH, dpi=150)
print(f"Saved decision-boundary figure to: {SAVE_FIG_PATH}")
plt.show()
```

```python
# Full pipeline: Linear Regression, Logistic Regression, Confusion Matrices, Decision Tree, Random Forest, SVM
# Copy into Colab / Jupyter and run. Adjust csv_path if needed.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score, roc_curve
)
from sklearn.metrics import ConfusionMatrixDisplay

import warnings
warnings.filterwarnings("ignore")

# -------------------------
# Configuration
# -------------------------
csv_path = '/content/drive/MyDrive/ML/Dataset of Diabetes Original .csv'  # update if needed
features = ['Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']
random_state = 11
test_size = 0.10

# Thresholds (you can change)
linear_threshold = 0.5      # used for Linear Regression -> binary
logistic_threshold = 0.5    # you can set 0.98 if you want same behavior as earlier

# -------------------------
# 1) Load & preprocess
# -------------------------
df = pd.read_csv(csv_path)
df = df.replace(['?', ' ', 'NA', 'nan', 'NaN', '--'], np.nan)

# Encode
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['CLASS_enc'] = df['CLASS'].apply(lambda x: 0 if x == 'N' else 1)

# Select X, y
X = df[features].copy()
y = df['CLASS_enc'].copy()

# Impute numeric missing values (mean)
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Scale (important for SVM; harmless for tree-based)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=test_size, random_state=random_state, stratify=y
)

# Small helper to evaluate & print results + show confusion matrix
def evaluate_model(name, y_train_true, y_train_pred, y_test_true, y_test_pred, y_test_score=None, show_cm=True):
    print(f"\n----- {name} -----")
    # Train metrics
    tr_acc = accuracy_score(y_train_true, y_train_pred)
    tr_prec = precision_score(y_train_true, y_train_pred, zero_division=0)
    tr_rec = recall_score(y_train_true, y_train_pred, zero_division=0)
    tr_f1 = f1_score(y_train_true, y_train_pred, zero_division=0)
    # Test metrics
    te_acc = accuracy_score(y_test_true, y_test_pred)
    te_prec = precision_score(y_test_true, y_test_pred, zero_division=0)
    te_rec = recall_score(y_test_true, y_test_pred, zero_division=0)
    te_f1 = f1_score(y_test_true, y_test_pred, zero_division=0)

    print(f"Training Accuracy: {tr_acc:.4f}")
    print(f"Training Precision: {tr_prec:.4f}")
    print(f"Training Recall: {tr_rec:.4f}")
    print(f"Training F1 Score: {tr_f1:.4f}\n")

    print(f"Testing Accuracy: {te_acc:.4f}")
    print(f"Testing Precision: {te_prec:.4f}")
    print(f"Testing Recall: {te_rec:.4f}")
    print(f"Testing F1 Score: {te_f1:.4f}\n")

    if show_cm:
        cm = confusion_matrix(y_test_true, y_test_pred)
        print("Confusion matrix (test):")
        print(cm)
        print("\nClassification report (test):")
        print(classification_report(y_test_true, y_test_pred, zero_division=0))
        # plot confusion matrix
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Diabetic','Diabetic'])
        disp.plot(cmap='Blues', values_format='d')
        plt.title(f"Confusion Matrix: {name}")
        plt.show()

    # if scores provided, compute ROC AUC
    if y_test_score is not None:
        try:
            auc = roc_auc_score(y_test_true, y_test_score)
            print(f"{name} Test ROC AUC: {auc:.4f}")
        except Exception:
            pass

# -------------------------
# 2) Linear Regression (as classifier via threshold)
# -------------------------
lin = LinearRegression()
lin.fit(X_train, y_train)
y_train_cont_lin = lin.predict(X_train)
y_test_cont_lin = lin.predict(X_test)
y_train_pred_lin = (y_train_cont_lin >= linear_threshold).astype(int)
y_test_pred_lin  = (y_test_cont_lin >= linear_threshold).astype(int)

evaluate_model("Linear Regression (threshold {:.2f})".format(linear_threshold),
               y_train, y_train_pred_lin, y_test, y_test_pred_lin,
               y_test_score=y_test_cont_lin)

# -------------------------
# 3) Logistic Regression
# -------------------------
log = LogisticRegression(max_iter=2000, random_state=42)
log.fit(X_train, y_train)
y_train_proba_log = log.predict_proba(X_train)[:,1]
y_test_proba_log  = log.predict_proba(X_test)[:,1]
y_train_pred_log = (y_train_proba_log >= logistic_threshold).astype(int)
y_test_pred_log  = (y_test_proba_log >= logistic_threshold).astype(int)

evaluate_model("Logistic Regression (threshold {:.2f})".format(logistic_threshold),
               y_train, y_train_pred_log, y_test, y_test_pred_log,
               y_test_score=y_test_proba_log)

# -------------------------
# 4) Decision Tree
# -------------------------
dt = DecisionTreeClassifier(max_depth=4, random_state=42)
dt.fit(X_train, y_train)
y_train_pred_dt = dt.predict(X_train)
y_test_pred_dt  = dt.predict(X_test)
# Trees can output probabilities too
y_test_proba_dt = dt.predict_proba(X_test)[:,1]

evaluate_model("Decision Tree (max_depth=4)", y_train, y_train_pred_dt, y_test, y_test_pred_dt, y_test_score=y_test_proba_dt)

# -------------------------
# 5) Random Forest
# -------------------------
rf = RandomForestClassifier(n_estimators=200, max_depth=8, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_train_pred_rf = rf.predict(X_train)
y_test_pred_rf  = rf.predict(X_test)
y_test_proba_rf = rf.predict_proba(X_test)[:,1]

evaluate_model("Random Forest (n_estimators=200, max_depth=8)", y_train, y_train_pred_rf, y_test, y_test_pred_rf, y_test_score=y_test_proba_rf)

# Print feature importances for Random Forest
rf_importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
print("\nRandom Forest feature importances (desc):")
print(rf_importances.round(4))

# -------------------------
# 6) SVM
# -------------------------
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)
y_train_pred_svm = svm.predict(X_train)
y_test_pred_svm  = svm.predict(X_test)
# probabilities (since probability=True)
y_test_proba_svm = svm.predict_proba(X_test)[:,1]

evaluate_model("SVM (RBF)", y_train, y_train_pred_svm, y_test, y_test_pred_svm, y_test_score=y_test_proba_svm)

# -------------------------
# 7) Summary table for test set
# -------------------------
summary = []
models_info = [
    ("Linear Regression", y_test_pred_lin, y_test_cont_lin),
    ("Logistic Regression", y_test_pred_log, y_test_proba_log),
    ("Decision Tree", y_test_pred_dt, y_test_proba_dt),
    ("Random Forest", y_test_pred_rf, y_test_proba_rf),
    ("SVM", y_test_pred_svm, y_test_proba_svm)
]

for name, y_pred, y_score in models_info:
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    try:
        auc = roc_auc_score(y_test, y_score)
    except Exception:
        auc = np.nan
    summary.append((name, acc, prec, rec, f1, auc))

summary_df = pd.DataFrame(summary, columns=['Model','Test Acc','Test Prec','Test Rec','Test F1','Test AUC']).set_index('Model')
print("\n=== Test set comparison ===")
display(summary_df.round(4))

# Optional: plot Test AUC/Accuracy bar charts
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.barplot(x=summary_df.index, y='Test Acc', data=summary_df.reset_index(), palette='viridis')
plt.title('Test Accuracy by Model')
plt.ylim(0,1)
plt.xticks(rotation=30)

plt.subplot(1,2,2)
sns.barplot(x=summary_df.index, y='Test AUC', data=summary_df.reset_index(), palette='magma')
plt.title('Test ROC-AUC by Model')
plt.ylim(0,1)
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python
# Full ML Pipeline: Linear, Logistic, Decision Tree, Random Forest, SVM
# Clean version with ONLY final test accuracy plot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_auc_score
)

import warnings
warnings.filterwarnings("ignore")

# -------------------------
# Configuration
# -------------------------
csv_path = '/content/drive/MyDrive/ML/Dataset of Diabetes Original .csv'  # Update if needed
features = ['Urea','Cr','HbA1c','Chol','TG','HDL','LDL','VLDL','BMI']
random_state = 11
test_size = 0.10

linear_threshold = 0.5
logistic_threshold = 0.5

# -------------------------
# 1) Load & preprocess
# -------------------------
df = pd.read_csv(csv_path)
df = df.replace(['?', ' ', 'NA', 'nan', 'NaN', '--'], np.nan)

# Encode
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['CLASS_enc'] = df['CLASS'].apply(lambda x: 0 if x == 'N' else 1)

# Select X, y
X = df[features].copy()
y = df['CLASS_enc'].copy()

# Impute missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Scale (important for SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=test_size, random_state=random_state, stratify=y
)

# Helper function for evaluation (no plotting)
def evaluate_model(name, y_train_true, y_train_pred, y_test_true, y_test_pred, y_test_score=None):
    tr_acc = accuracy_score(y_train_true, y_train_pred)
    tr_prec = precision_score(y_train_true, y_train_pred, zero_division=0)
    tr_rec = recall_score(y_train_true, y_train_pred, zero_division=0)
    tr_f1 = f1_score(y_train_true, y_train_pred, zero_division=0)

    te_acc = accuracy_score(y_test_true, y_test_pred)
    te_prec = precision_score(y_test_true, y_test_pred, zero_division=0)
    te_rec = recall_score(y_test_true, y_test_pred, zero_division=0)
    te_f1 = f1_score(y_test_true, y_test_pred, zero_division=0)

    print(f"\n----- {name} -----")
    print(f"Training Accuracy: {tr_acc:.4f} | Precision: {tr_prec:.4f} | Recall: {tr_rec:.4f} | F1: {tr_f1:.4f}")
    print(f"Testing Accuracy : {te_acc:.4f} | Precision: {te_prec:.4f} | Recall: {te_rec:.4f} | F1: {te_f1:.4f}")

    if y_test_score is not None:
        try:
            auc = roc_auc_score(y_test_true, y_test_score)
            print(f"{name} Test ROC AUC: {auc:.4f}")
        except Exception:
            auc = np.nan
    else:
        auc = np.nan
    return te_acc, te_prec, te_rec, te_f1, auc

# -------------------------
# 2) Linear Regression
# -------------------------
lin = LinearRegression()
lin.fit(X_train, y_train)
y_train_cont_lin = lin.predict(X_train)
y_test_cont_lin = lin.predict(X_test)
y_train_pred_lin = (y_train_cont_lin >= linear_threshold).astype(int)
y_test_pred_lin  = (y_test_cont_lin >= linear_threshold).astype(int)

acc_lin, prec_lin, rec_lin, f1_lin, auc_lin = evaluate_model(
    "Linear Regression", y_train, y_train_pred_lin, y_test, y_test_pred_lin, y_test_cont_lin
)

# -------------------------
# 3) Logistic Regression
# -------------------------
log = LogisticRegression(max_iter=2000, random_state=42)
log.fit(X_train, y_train)
y_train_proba_log = log.predict_proba(X_train)[:,1]
y_test_proba_log  = log.predict_proba(X_test)[:,1]
y_train_pred_log = (y_train_proba_log >= logistic_threshold).astype(int)
y_test_pred_log  = (y_test_proba_log >= logistic_threshold).astype(int)

acc_log, prec_log, rec_log, f1_log, auc_log = evaluate_model(
    "Logistic Regression", y_train, y_train_pred_log, y_test, y_test_pred_log, y_test_proba_log
)

# -------------------------
# 4) Decision Tree
# -------------------------
dt = DecisionTreeClassifier(max_depth=4, random_state=42)
dt.fit(X_train, y_train)
y_train_pred_dt = dt.predict(X_train)
y_test_pred_dt  = dt.predict(X_test)
y_test_proba_dt = dt.predict_proba(X_test)[:,1]

acc_dt, prec_dt, rec_dt, f1_dt, auc_dt = evaluate_model(
    "Decision Tree", y_train, y_train_pred_dt, y_test, y_test_pred_dt, y_test_proba_dt
)

# -------------------------
# 5) Random Forest
# -------------------------
rf = RandomForestClassifier(n_estimators=200, max_depth=8, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
y_train_pred_rf = rf.predict(X_train)
y_test_pred_rf  = rf.predict(X_test)
y_test_proba_rf = rf.predict_proba(X_test)[:,1]

acc_rf, prec_rf, rec_rf, f1_rf, auc_rf = evaluate_model(
    "Random Forest", y_train, y_train_pred_rf, y_test, y_test_pred_rf, y_test_proba_rf
)

# -------------------------
# 6) SVM
# -------------------------
svm = SVC(kernel='rbf', probability=True, random_state=42)
svm.fit(X_train, y_train)
y_train_pred_svm = svm.predict(X_train)
y_test_pred_svm  = svm.predict(X_test)
y_test_proba_svm = svm.predict_proba(X_test)[:,1]

acc_svm, prec_svm, rec_svm, f1_svm, auc_svm = evaluate_model(
    "SVM", y_train, y_train_pred_svm, y_test, y_test_pred_svm, y_test_proba_svm
)

# -------------------------
# 7) Final Summary Table
# -------------------------
summary = pd.DataFrame({
    'Model': ['Linear Regression', 'Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM'],
    'Test Accuracy': [acc_lin, acc_log, acc_dt, acc_rf, acc_svm],
    'Precision': [prec_lin, prec_log, prec_dt, prec_rf, prec_svm],
    'Recall': [rec_lin, rec_log, rec_dt, rec_rf, rec_svm],
    'F1 Score': [f1_lin, f1_log, f1_dt, f1_rf, f1_svm],
    'AUC': [auc_lin, auc_log, auc_dt, auc_rf, auc_svm]
}).set_index('Model')

print("\n=== Final Test Results ===")
display(summary.round(4))

# -------------------------
# 8) Final Plot: Test Accuracy Only
# -------------------------
plt.figure(figsize=(8,5))
sns.barplot(x=summary.index, y='Test Accuracy', data=summary.reset_index(), palette='coolwarm')
plt.title('Model Comparison — Test Accuracy', fontsize=14, fontweight='bold')
plt.ylabel('Test Accuracy')
plt.xlabel('Model')
plt.ylim(0,1)
plt.xticks(rotation=20)
for i, v in enumerate(summary['Test Accuracy'].values):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center')
plt.tight_layout()
plt.show()
```
